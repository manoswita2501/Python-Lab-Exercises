```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
data = pd.read_csv('UScereal.csv')
```

```python
# Bar Charts
# Two Variables
sns.barplot(x='mfr', y='calories', data=data, color='red')
plt.title('Bar Chart: Manufacturer vs Calories')
plt.show()
```

```python
# Three Variables
sns.barplot(x='mfr', y='calories', hue='shelf', data=data)
plt.title('Bar Chart: Manufacturer vs Calories with Shelf')
plt.show()
```

```python
# Scatter Plots
# Two Variables
plt.scatter(data['protein'], data['carbo'], color='darkorchid')
plt.title('Scatter Plot: Protein vs Carbohydrates')
plt.xlabel('Protein')
plt.ylabel('Carbohydrates')
plt.show()
```

```python
# Three Variables
sns.scatterplot(x='protein', y='carbo', hue='shelf', data=data)
plt.title('Scatter Plot: Protein vs Carbohydrates with Shelf')
plt.show()
```

```python
# Histograms
# Two Variables
plt.hist(data['calories'], bins=20, color='hotpink', edgecolor='black')
plt.title('Histogram: Calories Distribution')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Three Variables
sns.histplot(x='calories', hue='mfr', data=data, bins=20, multiple='stack')
plt.title('Histogram: Calories Distribution by Manufacturer')
plt.show()
```

```python
# Pie Chart
plt.pie(data['shelf'].value_counts(), labels=data['shelf'].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart: Shelf Distribution')
plt.show()
```

```python
# Pair Plot
sns.pairplot(data[['calories', 'protein', 'carbo']])
plt.suptitle('Pair Plot: Calories, Protein, Carbohydrates')
plt.show()
```

```python
# Bubble Plot
plt.scatter(data['protein'], data['carbo'], s=data['calories']*2, alpha=0.5, color='salmon')
plt.title('Bubble Plot: Protein vs Carbohydrates with Calories')
plt.xlabel('Protein')
plt.ylabel('Carbohydrates')
plt.show()
```

```python
# Box Plot
sns.boxplot(x='mfr', y='calories', data=data, color='fuchsia')
plt.title('Box Plot: Manufacturer vs Calories')
plt.show()
```

```python
# Violin Plot
sns.violinplot(x='mfr', y='calories', data=data, color='chartreuse')
plt.title('Violin Plot: Manufacturer vs Calories')
plt.show()
```

```python
# Area Plot
data[['protein', 'carbo']].plot.area(stacked=False)
plt.title('Area Plot: Protein vs Carbohydrates')
plt.show()
```

```python
# Any other plot (Heatmap)
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
